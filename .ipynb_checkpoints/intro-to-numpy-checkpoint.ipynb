### Dot products

```python
np.random.seed(0)
mat1 = np.random.randint(10, size=(5,3))
mat2 = np.random.randint(10, size=(5,3))
mat1
```

```python
mat2
```

```python
# This is ELEMENT-wise multiplication
mat1 * mat2
```

```python
# Another form of multiplication is a dotpruduct
np.dot(mat1, mat2)
```

```python
mat1.T
```

```python
np.dot(mat1.T, mat2)
```

```python
# Useful example:
np.random.seed(0)
# Number of jars sold
sales_amounts = np.random.randint(20, size=(5,3))
sales_amounts
```

```python
weekly_sales = pd.DataFrame(sales_amounts,
                           index=["Mon", "Tues", "Weds", "Thurs", 'Fri'],
                           columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
weekly_sales
```

```python
# Prices array
prices = np.array([10, 8, 12])
```

```python
butter_prices = pd.DataFrame(prices.reshape(1,3),
                            index=['Price'],
                            columns=['Almond Butter', 'Peanut Butter', 'Cashew Butter'])
butter_prices
```

```python
total_sales = np.dot(weekly_sales, butter_prices)
```

```python
butter_prices.shape, weekly_sales.shape
```

```python
butter_prices.shape, weekly_sales.T.shape
```

```python
total_sales = np.dot(butter_prices, weekly_sales.T)
total_sales
```

```python
np.dot(weekly_sales, butter_prices.T)
```

```python
total_sales_df = pd.DataFrame(total_sales,
                             index=['Total'],
                             columns=["Mon", "Tues", "Weds", "Thurs", 'Fri'])
total_sales_df
```

```python
weekly_sales['Total Sales'] = total_sales_df.iloc[0]
weekly_sales['Total Sales ($)'] = total_sales_df.T
```

```python
weekly_sales.drop(['Total Sales'], axis=1, inplace=True)
weekly_sales
```
