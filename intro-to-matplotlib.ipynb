### Which to use, pylot vs matplotlib OO?
* When doing a quick plot, pyplot method is simple and quick
* when plotting more advanced, use the OO method

```python
over_50 = heart_disease[heart_disease['age'] > 50]
over_50
```

```python
# pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method mixed with pyplot method
fig, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax)
# ax.set_xlim([45, 100])
```

```python
# OO method from scratch (most flexible)
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])

ax.set(title='Heart Disease and Cholesterol Levels',
       xlabel='Age',
       ylabel='Cholesterol Level')

ax.legend(*scatter.legend_elements(), title='Has Heart Disease?')

ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

```python
fig, ax = plt.subplots(figsize=(10, 6))

scatter = ax.scatter(x=over_50['age'],
          y=over_50['thalach'],
          c=over_50['target'])

ax.set(title='Heart Disease and Heart Rate',
      xlabel='Age',
      ylabel='Max Heart Rate')

ax.legend(*scatter.legend_elements(), title='Has Heart Disease?')

ax.axhline(over_50['thalach'].mean(),
          linestyle='--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol')
ax0.legend(*scatter.legend_elements(), title='Has Heart Disease')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

scatter_hr = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'])

ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate')
ax1.legend(*scatter_hr.legend_elements(), title='Has Heart Disease')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```

```python
plt.style.use('seaborn')

fig, (ax0, ax1) = plt.subplots(nrows=2,
                              ncols=1,
                              figsize=(10,10),
                              sharex=True)

scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

ax0.set(title='Heart Disease and Cholesterol Levels',
       ylabel='Cholesterol',
       xlim=[50, 80],
       ylim=[0, 600])
ax0.legend(*scatter.legend_elements(), title='Has Heart Disease')
ax0.axhline(y=over_50['chol'].mean(), linestyle='--')

scatter_hr = ax1.scatter(x=over_50['age'],
                        y=over_50['thalach'],
                        c=over_50['target'],
                        cmap='winter')

ax1.set(title='Heart Disease and Max Heart Rate',
       xlabel='Age',
       ylabel='Max Heart Rate',
       ylim=[50, 250])
ax1.legend(*scatter_hr.legend_elements(), title='Has Heart Disease')
ax1.axhline(y=over_50['thalach'].mean(), linestyle='--')

fig.suptitle('Heart Disease Analysis', fontsize=16, fontweight='bold')
```
